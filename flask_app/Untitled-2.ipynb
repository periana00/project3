```python
cur.execute('''
select movieid as id, avg(score) as score
from comments
group by movieid
''')
tmp = cur.fetchall()
df = pd.DataFrame(tmp)
df.columns = ['id', 'score']
```

```python
cur.execute('''
select name, movie_id as id from movie_genre
order by genre_id
''')
tmp = cur.fetchall()
df_genre = pd.DataFrame(tmp)
```

```python
df_genre.columns=['genre', 'id']
```

```python
tmp = pd.crosstab(df_genre.id, df_genre.genre)
```

```python
tmp.columns.name = None
tmp.sort_index(axis=1)
```

```python
df.set_index('id', inplace=True)
```

```python
tmp
```

```python
data = pd.merge(left=tmp, right=df, left_index=True, right_index=True)
```

```python
data
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(columns='score'), data['score'])
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
